```python
def toNodeChilds(childs ,parent ):
    nodeChilds = []
    for i in range(len(childs)):
        nodeChilds.append(Node(childs[i][0] , childs[i][1] , parent , parent.depth + 1 , parent.path_cost + 1))
    return nodeChilds
```

```python
class Node:
    def __init__(self , puzzle,prev , parent = None,depth = 0 , path_cost = 0):
        self.key = puzzle
        self.empty = self.key.index(0)
        self.previous_movement = prev
        self.parent = parent
        self.depth = depth
        self.path_cost = path_cost
        self.heuristic = 0
        self.fn = self.heuristic + self.path_cost
    def set_childs(self):
        list_temp = []
        if self.empty > 2 and self.previous_movement != 'D':
            temp = self.key.copy()
            temp[self.empty] , temp[self.empty -3] = temp[self.empty - 3] , temp[self.empty]
            list_temp.append((temp, 'U'))
        if self.empty < 6 and self.previous_movement != 'U':
            temp = self.key.copy()
            temp[self.empty] , temp[self.empty + 3 ] = temp[self.empty + 3] , temp[self.empty]
            list_temp.append((temp , 'D'))
        if self.empty % 3 != 2 and self.previous_movement != 'L':
            temp = self.key.copy()
            temp[self.empty] , temp[self.empty + 1] = temp[self.empty+1] , temp[self.empty]
            list_temp.append((temp , 'R'))
        if self.empty % 3 != 0 and self.previous_movement != 'R':
            temp = self.key.copy()
            temp[self.empty] , temp[self.empty - 1] = temp[self.empty -1 ] , temp[self.empty]
            list_temp.append((temp , 'L'))
        self.childs = toNodeChilds(list_temp , self)
    def manhattan_dist(self):
        manhat_sum = 0
        for i in range(1 , 9):
            i_location = self.key.index(i)
            index_difference = abs( i - i_location - 1)
            manhat_sum += (index_difference // 3) + (index_difference % 3)
        self.heuristic = manhat_sum
        self.fn = self.heuristic + self.path_cost
    def get_childs(self):
        return self.childs
    def get_prev(self):
        return self.previous_movement
    def get_puzzle(self):
        return self.key
    def get_parent(self):
        return self.parent
    def get_depth(self):
        return self.depth
```

```python
class Tree:
    def __init__(self , puzzle):
        self.goal = [1,2,3,4,5,6,7,8,0]
        self.puzzle = puzzle
        self.root = Node(self.puzzle , 'N')        
    def DFS(self):
        start = time.time()
        stack = [self.root]
        while stack != []:
            popped = stack.pop()
            popped.set_childs()
            if popped.get_puzzle() == self.goal:
                end = time.time()
                moves = []
                it = popped
                while it != None:
                    moves.append((it.key , it.previous_movement))
                    it = it.get_parent()
                for j in range(len(moves) -1 , -1 , -1):
                    print(moves[j][0])
                print('Action Sequence is : ' , end='')
                act_seq = ''
                for j in range(len(moves)-2 , -1 , -1):
                    print(moves[j][1] , end= '')
                    act_seq += moves[j][1]
                print()
                print(f'The time of running DFS algorithm is : {(end - start) * 1000:6f}ms')
                return (act_seq , (end - start)*1000)
            if popped.get_depth() < 22:
                stack.extend(popped.get_childs())
            
    def IDS(self):
        start = time.time()
        for i in range(33):
            stack = [self.root]
            
            while stack != []:
                popped = stack.pop()
                if popped.get_puzzle() == self.goal:
                    end = time.time()
                    moves = []
                    it = popped
                    while it != None:
                        moves.append((it.key , it.previous_movement))
                        it = it.get_parent()
                    for j in range(len(moves) -1 , -1 , -1):
                        print(moves[j][0])
                    print('Action Sequence is : ' , end='')
                    act_seq = ''
                    for j in range(len(moves)-2 , -1 , -1):
                        print(moves[j][1] , end= '')
                        act_seq += moves[j][1]
                    print()
                    print(f'The time of running IDS algorithm is : {(end - start) * 1000:6f}ms')
                    return (act_seq , (end - start)*1000)
                if popped.get_depth() < i:
                    stack.extend(popped.get_childs())
                
    def BFS(self):
        start = time.time()
        queue = [self.root]
        while queue != []:
            popped = queue.pop(0)
            popped.set_childs()
            if popped.get_puzzle() == self.goal:
                moves =  [] 
                it = popped
                while it != None:
                    moves.append((it.key , it.previous_movement))
                    it = it.get_parent()
                for j in range(len(moves)-1 , -1 , -1):
                    print(moves[j][0])
                print('Action Sequence is : ' , end='')
                act_seq = ''
                for j in range(len(moves)-2 , -1 , -1):
                    print(moves[j][1] , end= '')
                    act_seq += moves[j][1]
                print()
                end = time.time()
                print(f"The time of running BFS algorithm is : {(end - start) * 1000:6f}ms")
                return (act_seq , (end - start)*1000)
            queue.extend(popped.get_childs())
    def A_star(self):
        start = time.time()
        queue = [self.root]
        while queue != []:
            popped = queue.pop(0)
            popped.set_childs()
            if popped.get_puzzle() == self.goal:
                end = time.time()
                moves =  [] 
                it = popped
                while it != None:
                    moves.append((it.key , it.previous_movement))
                    it = it.get_parent()
                for j in range(len(moves)-1 , -1 , -1):
                    print(moves[j][0])
                print('Action Sequence is : ' , end='')
                act_seq = ''
                for j in range(len(moves)-2 , -1 , -1):
                    print(moves[j][1] , end= '')
                    act_seq += moves[j][1]
                print()
                print(f"The time of running A* algorithm is : {(end - start) * 1000:6f}ms")
                return [act_seq , (end - start)*1000]
            for child in popped.get_childs():
                child.manhattan_dist()
                i = 0
                while i < len(queue) and child.fn > queue[i].fn:
                    i += 1
                queue.insert(i , child)
    def UCS(self):
        start = time.time()
        queue = [self.root]
        while queue != []:
            popped = queue.pop(0)
            popped.set_childs()
            if popped.get_puzzle() == self.goal:
                end = time.time()
                moves =  [] 
                it = popped
                while it != None:
                    moves.append((it.key , it.previous_movement))
                    it = it.get_parent()
                for j in range(len(moves)-1 , -1 , -1):
                    print(moves[j][0])
                print('Action Sequence is : ' , end='')
                act_seq = ''
                for j in range(len(moves)-2 , -1 , -1):
                    print(moves[j][1] , end= '')
                    act_seq += moves[j][1]
                print()
                print(f"The time of running UCS algorithm is : {(end - start) * 1000:6f}ms")
                return 
            for child in popped.get_childs():
                i = 0
                while i < len(queue) and child.fn > queue[i].fn:
                    i += 1
                queue.insert(i , child)
```
